```python
import pymongo 
import pandas as pd 
df = pd.read_csv("twitter-stocks.csv")
df.head(10)
```

```python
# lets convert it into dict 
data_dict = df.to_dict("records")
```

```python
import os
project_name = "twitterSTock"
collection_name = "twitterSTockCollection"
collection_url = os.getenv("MONGO_DB_KEYS") 
```

```python
# now connect to the mongo db and create a database and collection and insert the data into it
client = pymongo.MongoClient(collection_url)
collection = client[project_name]
name = collection[collection_name]
try:
    rec = name.insert_many(data_dict)
    print(f"Successfully inserted {len(rec)} documents")
except Exception as e:
    print(f"Error inserting data: {e}")
```

```python
df = pd.DataFrame(list(name.find()))
```
